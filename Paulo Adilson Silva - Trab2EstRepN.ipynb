### UTILIZE **DUAS** ESTRURUAS DE REPETIÇÃO. Criar um programa para realizar um levantamento nas salas de aula do curso de ADS dos 6 módulos, mas não é conhecida a quantidade de alunos das mesmas, para isso utilize a idade com valor zero para sair da repetição interna. Calcule e apresente:​

1. Quantidade de homens e mulheres de cada sala de aula;​

2. Quantidade de pessoas de cada sala de aula com idade acima e abaixo de 18 anos;​

3. Média de idades de cada sala de aula;​

4. Média de idades do curso todo;​

5. Quantidade de homens e mulheres do curso todo.​

```
hTot = nTot = sId = 0
for sala in range(6):
    print('\nInforme a idade e o sexo dos alunos da {}ª sala.' .format(sala+1))
    idade = int(input('\nInforme a idade ou digite 0 para sair: '))
    gt18 = hTotS = sIdS = nTotS = 0
    while idade > 0:
        sexo = input('Informe o sexo do(a) aluno(a) (m ou f): ').lower()    
        if idade >= 18:
            gt18 += 1
        if sexo == 'm':
            hTotS += 1
            hTot +=1
        nTotS +=1
        nTot +=1
        sId += idade
        sIdS += idade
        idade = int(input('\nInforme a idade ou digite 0 para sair: '))
    print('\nA quantidade de pessoas menores e maiores de 18 anos da {}ª sala é {} e {}, respectivamente.' .format(sala+1,nTotS-gt18,gt18))
    print('A quantidade de homens e mulheres da {}ª sala é {} e {}, respectivamente.' .format(sala+1,hTotS,(nTotS-hTotS)))
    print('A média de idade da {}ª sala é {:.2f} anos.' .format(sala+1,sIdS/nTotS))
print('\nA quantidade total de pessoas cursando ADS é {}.'.format(nTot))
print('A média de idade do curso de ADS é {:.2f} anos.' .format(sId/nTot))
print('A quantidade de homens e mulheres do curso de ADS é {} e {}, respectivamente.' .format(hTot,nTot-hTot))
print('A porcentagem de homens e mulheres do curso de ADS é {:.2f}% e {:.2f}%, respectivamente.' .format(hTot/nTot*100,(nTot-hTot)/nTot*100))
```
