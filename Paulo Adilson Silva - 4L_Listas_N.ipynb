### 10. Um vetor é palíndromo se ele não se alterar quando o mesmo for invertido. Por exemplo, o vetor original vo = {1, 3, 5, 2, 2, 5, 3, 1} é palíndromo, pois ele invertido é vi = {1, 3, 5, 2, 2, 5, 3, 1}, igual ao original. Escreva um algoritmo que verifique se um vetor é palíndromo, fazendo comparação de posição por posição do vetor origem (vo) com o vetor invertido(vi).


```
# 10. Um vetor é palíndromo se
lo = []
qtN = int(input('Quantos números pretende verificar se são palíndromo? '))
for n in range(qtN):
    lo.append(int(input('Informe o {}º número :' .format(n+1))))
li = lo[::-1]
somaL = 0
for i in range(len(lo)):
    if lo[i] == li[i]:
        somaL += 1
if somaL == len(lo):
    print('\nA lista é um palíndromo!')
else:
  print('\nA lista NÃO é um palíndromo!')
```

```
# ALTERNATIVO 10. Um vetor é palíndromo se
lo = []
qtN = int(input('Quantos números pretende verificar se são palíndromo? '))
for n in range(qtN):
    lo.append(int(input('Informe o {}º número :' .format(n+1))))
li = lo[::-1]
somaL = 0
for i in range(len(lo)):
    if lo[i] == li[i]:
        somaL += 1
    else:
        print('\nA lista NÃO é um palíndromo!')
        break
if somaL == len(lo):
    print('\nA lista é um palíndromo!')
```

### 24. Faça um programa que:
* leia uma lista com quinze posições para números inteiros.
* Crie, a seguir, um lista resultante que contenha todos os números primos do lista digitado.
* Escreva o lista resultante.

```
# 24. Faça um programa que:
l = []
for n in range(15):
  l.append(int(input('Informe um número inteiro: ')))
lPrimos = []
qtDiv = 0
for n in range(len(l)):
  for d in range(1,n+1):
    if l[n] % d == 0:
      qtDiv += 1
  if qtDiv == 2:
    lPrimos.append(n)
  qtDiv = 0
print('\nOs números primos desta lista de números são: ', end='')
print(*lPrimos, sep=', ', end='.')
```

### 25. Faça um programa que:
* leia uma lista com quinze posições para números inteiros.
* Depois da leitura, divida todos os seus elementos pelo maior valor do lista.
* Mostre o lista após os cálculos.

```
# 25. Faça um programa que:
l = []
for n in range(5):
  l.append(int(input('Informe um número inteiro: ')))
  c = 0
  while c < len(l):
    if c == 0:
      maN = l[c]
    else:
      if maN < l[c]:
        maN = l[c]
    c += 1
for n in range(len(l)):
  #l[n] /= maN
  l[n] = round((l[n]/maN),3)
  round(l[n],4)
print('\nOs valores da lista dividida pelo maior valor deles gera a seguinte lista: ', end='')
print(*l, sep=', ', end='.')
```
