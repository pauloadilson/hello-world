### 5. Faça um programa que receba duas notas de seis alunos. Calcule e mostre:

* a média aritmética das duas notas de cada aluno; e
* a mensagem que está na tabela a seguir:


Média aritimética | Mesagem
---|---
Até 3|Reprovado
Entre 3 e 7 |Exame
De 7 para cima| Aprovado

* o total de alunos aprovados;
* o total de alunos de exame;
* o total de alunos reprovados;

````
# 5. Faça um programa que receba duas notas de seis alunos.
nroAlunos = 0
totReprov = 0
totExame = 0
totAprov = 0
mediaTot = 0
while nroAlunos < 6:
  nota1 = float(input('\nInforme a primeira nota do {0}º aluno: ' .format(nroAlunos+1)))
  nota2 = float(input('Informe a segunda nota do {0}º aluno: ' .format(nroAlunos+1)))
  media = (nota1 + nota2) / 2
  if media <= 3:
    classif = 'reprovado'
    totReprov += 1
  elif media < 7:
    classif = 'em exame'
    totExame += 1
  else:
    classif = 'aprovado'
    totAprov += 1
  print('A média do {0}º aluno é {1:.1f} e ele está {2}.' .format(nroAlunos+1,media,classif))
  mediaTot = mediaTot + media
  nroAlunos += 1
print('\nO total de alunos aprovados é {0}.' .format(totAprov))
print('O total de alunos em exame é {0}.' .format(totExame))
print('O total de alunos reprovados é {0}.' .format(totReprov))
print('A média da classe é {0:.1f}.' .format(mediaTot/6))
  
````
