### 1. Faça um programa que carregue uma matriz 2 x 2,  que representa o salário de 4 funcionários, calcule e mostre a soma total de todos os elementos que será o montante pago pela empresa a esses funcionários

```
#1. Faça um programa que carregue uma matriz 2 x 2, que representa o salário de 4 funcionários, calcule e mostre a soma total de todos os elementos que será o montante pago pela empresa a esses funcionários
import numpy as np
sal = np.empty([2,2])
soma = 0
for l in range(2):
  for c in range(2):
    sal[l,c] = float(input('Informe o salário do funcionário: R$'))
    soma += sal[l,c]
#fPag = np.sum(sal)
print('O montante total pago pela empresa foi R$ {:.2f}.' .format(soma))
```

###7.Faça um programa que carregue uma matriz 2 x 2 int e encontre o maior elemento da matriz, crie uma matriz resultante que armazene e mostre o cálculo dos elementos da matriz original multiplicados pelo maior valor encontrado.


```
# 7.Faça um programa que carregue uma matriz 2 x 2 int
import numpy as np
mat = np.empty([2,2], dtype=int)
mat2 = np.empty([2,2], dtype=int)
for l in range(2):
  for c in range(2):
    mat[l,c] = int(input('Informe um número inteiro: '))
    if l == 0 and c == 0:
      maN = mat[l,c]
    else:
      if maN < mat[l,c]:
        maN = mat[l,c]
print()
for l in range(2):
  for c in range(2):
    mat2[l,c] = mat[l,c]*maN
    print('{} x {} = {}.' .format(mat[l,c],maN,mat2[l,c]))
print('\n A matriz multiplicada pelo maior número é: ', end='')
print(*mat2, sep=", ", end=".")
```

###8. Faça um programa que a partir de uma matriz 4x4 mostre a soma dos elementos da diagonal principal. Mostre também  a soma dos elementos que estão acima da diagonal principal.


```
# 8. Faça um programa que a partir de uma matriz 4x4
import numpy as np
mat = np.empty([4,4], dtype=int)
sDiag = 0
sAcDiag = 0
for l in range(4):
  for c in range(4):
    mat[l,c] = int(input('Informe um número inteiro: '))
    if l == c:
      sDiag += mat[l,c]
    elif c > l:
      sAcDiag += mat[l,c]
print('\nA soma dos elementos da diagnonal principal é {}.' .format(sDiag))
print('A soma dos elementos ACIMA da diagnonal principal é {}.' .format(sAcDiag))
```

###9. Faça um programa que carregue uma matriz 10 x 20 com números inteiros e some cada uma das linhas, armazenando o resultado das somas em um vetor. A seguir, multiplique cada elemento da matriz pela soma da linha e mostre a matriz resultante.

###10. Faça um programa que carregue: *um vetor com oito posições com os nomes das lojas; * um outro vetor com quatro posições com os nomes dos produtos; * uma matriz com os preços de todos os produtos em cada loja. O programa deve mostrar todas as relações (nome do produto – nome da loja) nas quais o preço não ultrapasse R$ 100,00.


```
# 10. Faça um programa que carregue:
import numpy as np
nL = []
nPro = []
prePL = np.empty([3,2], dtype=float)
for lo in range(3):
  nL.append(input('Informe o nome da {}ª loja: ' .format(lo+1)))
print()
for pro in range(2):
  nPro.append(input('Informe o nome do {}º produto: ' .format(pro+1)))
print()
for l in range(len(nL)):
  for c in range(len(nPro)):
    prePL[l,c] = float(input("Informe o preço do produto '{}' da loja '{}': R$" .format(nPro[c],nL[l])))
  print()
for l in range(len(nL)):
  for c in range(len(nPro)):
    if prePL[l,c] < 100:
      print("O preço do produto '{}' da loja '{}' não ultrapassa R$100,00, custando R${:.02f}.".format(nPro[c],nL[l],prePL[l,c]))
```
